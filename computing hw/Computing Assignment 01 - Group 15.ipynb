# Part 01 Object Store

```python
import csv
# define a class object_store to store data
class object_store:
    def __init__(self,filename):
        self.store = []
        self.list = []
        self.attribute=[]
        # counter to count the number of L
        counter = 0
        with open(filename,'r')as file:
            reader = csv.reader(file)
            # read each line
            for i in reader:
                j = 0
                while j < len(i):
                    if len(i[j]) != 0:
                        j = j + 1
                    # remove the empty data
                    else:
                        i.remove(i[j])
                # use this variable to judge whether this is contained in Li 
                flag = 0
                for j in range(len(self.list)):
                    # if it is contained
                    if i[0] == self.list[j]:
                        # add information
                        self.store[j].append([None,i[0],[]])
                        for k in range(1,(len(i)-1)//2+1):
                            self.store[j][len(self.store[j])-1][2].append([None,i[2*k-1],i[2*k]])
                            if i[2*k-1] not in self.attribute:
                                self.attribute.append(i[2*k-1])
                        flag=1
                        break
                # if not contained        
                if flag == 0:
                    # add information
                    self.store.append([])
                    self.store[len(self.list)].append([None,i[0],[]])
                    for k in range(1,(len(i)-1)//2+1):
                        self.store[len(self.list)][len(self.store[len(self.list)])-1][2].append([None,i[2*k-1],i[2*k]])
                        if i[2*k-1] not in self.attribute:
                                self.attribute.append(i[2*k-1])
                    self.list.append(i[0]) 
            # add identifier
            id = 1
            for i in range(len(self.store)):
                for j in range(len(self.store[i])):
                    self.store[i][j][0] = id
                    id += 1
                    for k in range(len(self.store[i][j][2])):
                        self.store[i][j][2][k][0] = id
                        id += 1
```

```python
class ENV_STACK():
    #The env stack in 2 stack machine
    #we input our object store as our parameter
    def __init__(self,object_store,size=20):
        self.items=[None]*size
        self.numItems = 0
        self.size=size
        self.object_store=object_store
        all_i=[]
        self.top=None
        for i in object_store.store:
            for j in i:
                all_i.append(j[0])
        self.push(all_i)
    #top function as before
    def top(self):
        if self.numItems!=0:
            return self.items[self.numItems-1]
        raise Error("stack is empty")
    #push function as before
    def push(self,item):
        if self.numItems ==self.size:
            self.allocate()
        self.items[self.numItems]=item
        self.top=self.items[self.numItems]
        self.numItems+=1
    #allocate function as before
    def allocate(self):
        newlength=2*self.size
        newStack=[None]*newlength
        for i in range(self.numItems):
            newStack[i]=self.items[i]
        self.items=newStack
        self.size=newlength
    #pop function as before
    def pop(self):
        if self.numItems==int(self.size/4):
            self.deallocate()
        if self.numItems!=0:           
            a=self.items[self.numItems-1] 
            self.items[self.numItems-1]=None
            self.numItems-=1
            if self.numItems!=0:
                self.top=self.items[self.numItems-1]
            return a
        raise Exception("stack is empty")
    #deallocate function as before
    def deallocate(self):
        newlength=int(self.size/2)
        newsStack=[None]*newlength
        for i in range(self.numItems):
            newStack[i]=self.items[i]
        self.items=newStack
        self.size=newlength
    #is_empty function as before
    def is_Empty(self):
        if self.numItems!=0:
            return False
        return True
    
    # n is the external name of storerage object, 
    # search will return us the list of identifiers    
############################this is the new function of env stack#######################################
    def search(self,n):
        res=[] #this is the list to store identifiers
        #the condition when n is L1,L2,L3...the list
        if n in self.object_store.list:
            for j in self.object_store.store:
                for k in j:
                    if n==k[1] and k[0] in self.top:
                        res.append(k[0])
        #the condition when n is cinema, title...the attribute name
        if n in self.object_store.attribute:
            for i in self.object_store.store:
                for j in i:
                    for k in j[2]:
                        if k[1]==n and k[0] in self.top:
                            res.append(k[0])
        if res==[]:
            return None
        return res
##########################################################################################################
class RES_STACK():
    def __init__(self,size=20):
        self.items=[None]*size
        self.numItems = 0
        self.size=size
        self.top=None
    def top(self):
        if self.numItems!=0:
            return self.items[self.numItems-1]
        raise Error("stack is empty")
    def push(self,item):
        if self.numItems ==self.size:
            self.allocate()
        self.items[self.numItems]=item
        self.top=self.items[self.numItems]
        self.numItems+=1
    def allocate(self):
        newlength=2*self.size
        newStack=[None]*newlength
        for i in range(self.numItems):
            newStack[i]=self.items[i]
        self.items=newStack
        self.size=newlength
    def pop(self):
        if self.numItems==int(self.size/4):
            self.deallocate()
        if self.numItems!=0:
            a=self.items[self.numItems-1] 
            self.items[self.numItems-1]=None
            self.numItems-=1
            if self.numItems!=0:
                self.top=self.items[self.numItems-1]
            return a
        raise Exception("stack is empty")
    def deallocate(self):
        newlength=int(self.size/2)
        newStack=[None]*newlength
        for i in range(self.numItems):
            newStack[i]=self.items[i]
        self.items=newStack
        self.size=newlength
    def is_Empty(self):
        if self.numItems!=0:
            return False
        return True
        
```

```python
#Part3
class Two_stack_machine:
    def __init__(self,Env,Res,object_store):
        self.env=Env
        self.res=Res       
        self.Object_store = object_store
        ##all list string
        self.attribute=object_store.attribute
        self.list=object_store.list+object_store.attribute
        ##string name of all command
        self.command=["where","x","."]
        self.one_command=["count", "sum", "min", "max", "average", "distinct", "not" , "Exists","deref","nested"]
        self.two_command=["=", "<", ">", "<=", ">=", "neq","and", "or","+", "-", "*", "/","Contains", "In"]
        self.value=0
        self.left=""
        self.whgo=0
        ##insert like this: count Performance where Theater . Title
    
        m=input("Input you command command:")
        self._eval(m)
        
        #self.operators contains all commands.When calling functions, search here first.
        #If not here, return ERROR
        self.operators = ["plus" , "minus" , "multiply" , "divide" , "dereference"]
  ########################################################################################################################            
    def _eval(self,string):
        #去括号
        if string=="":
            return
    
        #空格分离
        m=string.split(' ',1)
        if m[0]=="(":
            count=1
            leng=len(m[1])
            temp=m[1].split(" ")
            if (temp[3:6]==[")", "where" ,"("])&(temp[1]=="x"):
                    self._eval(temp[0])
                    k=m[1].split(" ",2)
                    self.X_equaljoin(k[2][0:])
                    return
            for i in range(0,leng):
                if (m[1][i]=="("):
                    count=count+1
                if (m[1][i]==")"):
                    count=count-1    
                    if count==0:
                        self._eval(m[1][:i-1])
                        self._eval(m[1][(i+2):])
                        return
                    
        #检查是不是一个List
        if m[0] in self.list:
            self.select(m[0])
            if len(m)!=1:
                self._eval(m[1])
            else:
                return
            
        #检查是不是 query类的操作
        elif m[0]=="where":
            self.where(m[1])
            self._eval(self.left)
            return
            
        elif "."in m[0][0]:
            self.getattribute(m[0][1:])
            if len(m)!=1:
                self._eval(m[1])
            return
        
        elif m[0]=="x":
            self.X(m[1])
            self._eval(self.left)
            return

            
        ##if a command with algebric    
        elif m[0] in self.two_command:
            self._eval(m[1])
            if m[0]=="=":  
                self.equal()
                return
            if m[0]=="<":
                self.smaller()
                return
            if m[0]==">":
                self.larger()
                return
            if m[0]=="<=":
                self.smaller_and_equal()
                return
            if m[0]==">=":
                self.larger_and_equal()
                return
            if m[0]=="neq":
                self.not_equal()
                return
            if m[0]=="and":
                self.And()
                return
            if m[0]=="or":
                self.Or()
                return
            if m[0]=="+":
                self.plus()
                return
            if m[0]=="-":
                self.minus()
                return
            if m[0]=="*":
                self.multiple()
                return
            if m[0]=="/":
                self.divide()
                return
            if m[0]=="Contains":
                self.Contains()
                return
            if m[0]=="In":
                self.In()
                return
        elif m[0] in self.one_command:
            self._eval(m[1])
            if m[0]=="Exists":
                self.Exist()
                return
            if m[0]=="count":
                self.count()
                return
            if m[0]=="sum":
                self.Sum()
                return
            if m[0]=="min":
                self.Min()
                return
            if m[0]=="max":
                self.Max()
                return
            if m[0]=="average":
                self.average()
                return
            if m[0]=="distinct":
                self.distinct()
                return
            if m[0]=="not":
                self.Not()
                return
            if m[0]=="deref":
                self.dereference()
                return
            if m[0]=="nested":
                self.nested()
                return
        else:
            self.value=1
            self.select(m[0])
            self.value=0
            if len(m)!=1:
                self.left=m[1]
                return 
        
  ########################################################################################################################    
    def where(self,obj_2):
        self.whgo=1
        result=[]
        #Performance 
        for i in self.res.top:
            self.env.push(self.nested(i))
            
            #eval the following part
            self._eval(obj_2)
            
            #if correct  
           
            if self.res.pop()==True:
                result.append(i)
            self.env.pop()
        self.res.pop()
        self.res.push(result) 
        self.whgo=0
        return  
   ##select function
    def select(self,name):
        if  self.value==0:
            f=self.env.search(name)
            self.res.push(f)
            if (name in self.attribute )& (self.whgo==1):
                self.dereference()
             
            return
        
        else:
            name=name.split()
            self.res.push(name)
             
            return
        
        
    ##the query Q.A
    def getattribute(self,attr):
        m=self.res.pop()
        
        self.env.push(self.nested(m[0]))
        self._eval(attr) 
        return
    
    ##the Cartesian Q1xQ2
    def X(self,item_2):
        result=[]
        a=[]
        b=[]
        m=self.res.pop()
        for i in m:
            a.append(self.nested(i))
        self._eval(item_2)
        m=self.res.pop()
        for i in m:
            b.append(self.nested(i))
        for i in a:
            for j in b:
                s=[]
                s.extend(i)
                s.extend(j)
                m=tuple(s)
                result.append(m)
        self.res.push(result)
        return
########################################################################################################################    
    # ( Q1 x Q2 ) where ( a = b )    
    def X_equaljoin(self,item_2):
        # part_1 is used to load Q1
        part_1=self.res.pop()
        part_remain=item_2.split(" ) where ( ")
        # part_2 is used to load Q2
        part_2=self.env.search(part_remain[0])
        part_remain_1=part_remain[1]
        part_remain_2=part_remain_1.split(" = ")
        # part_3 is used to load a
        part_3=part_remain_2[0]
        # part_4 is used to load b
        part_4=part_remain_2[1][:-2]
        # result_1 is a list to load all Q1's value of a     
        result_1 = []
        for i in part_1:
            # mark is a variable to show whether we have find the target to help us exit efficiently
            mark=0
            if i > self.Object_store.store[-1][-1][-1][-1][0]:
                raise Exception("It's out of range")
            for j in self.Object_store.store:
                for k in j:
                    if i == k[0]:       
                        for loc in range(len(k[2])):
                            if k[2][loc][1] == part_3:
                                result_1.append(k[2][loc][2])
                                mark = 1
                                break
                        if mark==0:
                            # this is for the special case that it doesn't contain a 
                            result_1.append(None)
                            mark=1
                        break
                if mark==1:
                    break
        # result_2 is a list to load all Q2's value of b
        result_2 = []            
        for i in part_2:
            # mark is a variable to show whether we have find the target to help us exit efficiently
            mark=0
            if i > self.Object_store.store[-1][-1][-1][-1][0]:
                raise Exception("It's out of range")
            for j in self.Object_store.store:
                for k in j:
                    if i == k[0]: 
                        for loc in range(len(k[2])):
                            if k[2][loc][1] == part_4:
                                result_2.append(k[2][loc][2])
                                mark = 1
                                break
                        if mark==0:
                            # this is for the special case that it doesn't contain b
                            result_2.append(None)
                            mark=1
                        break
                if mark==1:
                    break
        # result is our true result of X_equaljoin
        result=[]
        for loc_1 in range(len(result_1)):
            if result_1[loc_1] != None:
                for loc_2 in range(len(result_2)):
                    if result_1[loc_1]==result_2[loc_2]:
                        # execute ( Q1 x Q2 ) where ( a = b ) 
                        thing=self.nested(part_1[loc_1])+self.nested(part_2[loc_2])
                        result.append(thing)
        # push back result
        self.res.push(result)   
############################################################################################################################
    def plus(self):      
        temp2 = self.res.pop()
        temp1 = self.res.pop()
        if type(temp1)==type(temp2):
            
            #judge whether obj1 and obj2 are of same type  
            temp1 = temp1 + temp2
            self.res.push(temp1) 
        elif type(temp1)==int and type(temp2)==list and len(temp2)==1 and type(temp2[0])==int:
            temp1 = temp1 + temp2
            self.res.push(temp1)     
        elif type(temp2)==int and type(temp1)==list and len(temp1)==1 and type(temp1[0])==int:
            temp1 = temp1 + temp2
            self.res.push(temp1) 
        else:             #if not, raise error
            raise Exception("Error occured in Plus") 
            
    def minus(self):
        temp2 = self.res.pop()
        temp1 = self.res.pop()
        if is_number(temp1) and is_number(temp2):
            #judge whether obj1 and obj2 are both digits  
            temp1 = temp1 - temp2
            self.res.push(temp1) 
        else:             #if not, raise error
            raise Exception("Error occured in Minus") 
            
    def multiple(self):
        temp1 = self.res.pop()
        temp2 = self.res.pop()
        if is_number(temp1) and is_number(temp2):
            #judge whether obj1 and obj2 are both digits  
            temp1 = temp1 * temp2
            self.res.push(temp1) 
        else:             #if not, raise error
            raise Exception("Error occured in Multiple")
            
    def divide(self): #res = obj1 / obj2
        temp2 = self.res.pop()
        temp1 = self.res.pop()
        if is_number(temp1) and is_number(temp2):   #judge whether obj1 and obj2 are both digits  
            if temp2 == 0:         #ensure obj2 is not 0
                raise Exception("Error occured in Divide")
            else:
                temp1 = temp1 / temp2
                self.res.push(temp1) 
        else:             #if not, raise error
            raise Exception("Error occured in Divide2")
            
############################################################################################

    def smaller(self):  #if obj1 is less than obj2, return True, else return false
        temp2 = self.res.pop()
        temp1 = self.res.pop()
        if type(temp1)==type(temp2): 
            if temp1 < temp2:
                self.res.push(True)
            else:
                self.res.push(False)
        elif type(temp1)==int and type(temp2)==list and len(temp2)==1 and type(temp2[0])==int:
            if temp1 < temp2:
                self.res.push(True)
            else:
                self.res.push(False)    
        elif type(temp2)==int and type(temp1)==list and len(temp1)==1 and type(temp1[0])==int:
            if temp1 < temp2:
                self.res.push(True)
            else:
                self.res.push(False)
        else:             #if not, raise error
            raise Exception("Cannot evaluate in python.")
        

    def smaller_and_equal(self):
        temp2 = self.res.pop()
        temp1 = self.res.pop()
        if type(temp1)==type(temp2): 
            if temp1 <= temp2:
                self.res.push(True)
            else:
                self.res.push(False)
        elif type(temp1)==int and type(temp2)==list and len(temp2)==1 and type(temp2[0])==int:
            if temp1 <= temp2:
                self.res.push(True)
            else:
                self.res.push(False)    
        elif type(temp2)==int and type(temp1)==list and len(temp1)==1 and type(temp1[0])==int:
            if temp1 <= temp2:
                self.res.push(True)
            else:
                self.res.push(False)
        else:             #if not, raise error
            raise Exception("Cannot evaluate in python.")      
            
    def larger(self):
        temp2 = self.res.pop()
        temp1 = self.res.pop()
        if type(temp1)==type(temp2): 
            if temp1 > temp2:
                self.res.push(True)
            else:
                self.res.push(False)
        elif type(temp1)==int and type(temp2)==list and len(temp2)==1 and type(temp2[0])==int:
            if temp1 > temp2:
                self.res.push(True)
            else:
                self.res.push(False)    
        elif type(temp2)==int and type(temp1)==list and len(temp1)==1 and type(temp1[0])==int:
            if temp1 > temp2:
                self.res.push(True)
            else:
                self.res.push(False)
        else:             #if not, raise error
            raise Exception("Cannot evaluate in python.")
            
    def larger_and_equal(self):
        temp2 = self.res.pop()
        temp1 = self.res.pop()
        if type(temp1)==type(temp2): 
            if temp1 >= temp2:
                self.res.push(True)
            else:
                self.res.push(False)
        elif type(temp1)==int and type(temp2)==list and len(temp2)==1 and type(temp2[0])==int:
            if temp1 >= temp2:
                self.res.push(True)
            else:
                self.res.push(False)    
        elif type(temp2)==int and type(temp1)==list and len(temp1)==1 and type(temp1[0])==int:
            if temp1 >= temp2:
                self.res.push(True)
            else:
                self.res.push(False)
        else:             #if not, raise error
            raise Exception("Cannot evaluate in python.")
        
    def equal(self):
        
        temp1 = self.res.pop()
        temp2 = self.res.pop()
        if type(temp1)==type(temp2):
            #judge whether obj1 and obj2 are of same type  
            if temp1 == temp2:
                self.res.push(True)
            else:
                self.res.push(False) 
        elif type(temp1)==int and type(temp2)==list and len(temp2)==1 and type(temp2[0])==int:
            if temp1 == temp2:
                self.res.push(True)
            else:
                self.res.push(False)     
        elif type(temp2)==int and type(temp1)==list and len(temp1)==1 and type(temp1[0])==int:
            if temp1 == temp2:
                self.res.push(True)
            else:
                self.res.push(False)
        else:
            raise Error("Cannot evaluate in python.")
        
    def not_equal(self):
        temp1 = self.res.pop()
        temp2 = self.res.pop()
        if type(temp1)==type(temp2):
            #judge whether obj1 and obj2 are of same type  
            if temp1 != temp2:
                self.res.push(True)
            else:
                self.res.push(False) 
        elif type(temp1)==int and type(temp2)==list and len(temp2)==1 and type(temp2[0])==int:
            if temp1 != temp2:
                self.res.push(True)
            else:
                self.res.push(False)     
        elif type(temp2)==int and type(temp1)==list and len(temp1)==1 and type(temp1[0])==int:
            if temp1 != temp2:
                self.res.push(True)
            else:
                self.res.push(False)
        else:
            raise Error("Cannot evaluate in python.")
############################################################################################
    def And(self):
        temp1 = self.res.pop()
        temp2 = self.res.pop()            
        if ((temp1 == True) or (temp1 == False)) and ((temp2== True) or (temp2 == False)):
            temp2 = temp1 and temp2
            self.res.push(temp2)
            return
        raise Exception("Error occured in And")
        
    def Or(self):
        temp1 = self.res.pop()
        temp2 = self.res.pop()
        if ((temp1 == True) or (temp1 == False)) and ((temp2== True) or (temp2 == False)):
            temp2 = temp1 or temp2
            self.res.push(temp2)   
            return
        raise Exception("Error occured in Or")
        
    def Not(self):
        temp1 = self.res.pop()
        if (temp1 == True) or (temp1 == False):
            temp1 = not(temp1)
            self.res.push(temp1)   
            return
        raise Exception("Error occured in Not")
############################################################################################  
    def Sum(self):
        temp = self.res.pop()
        Sum = 0
        for i in temp:
            if is_number(i):
                Sum += i
            else:
                raise Exception("Cannot operate sum")
        self.res.push(Sum)    #push back the average number  
        
    def Min(self):
        temp = self.res.pop()
        for i in temp:
            if is_number(i):
                continue
            else:
                raise Exception("Cannot operate Min")
        Min = min(temp)
        self.res.push(Min)    #push back the minimum
        
    def Max(self):
        temp = self.res.pop()
        for i in temp:
            if is_number(i):
                continue
            else:
                raise Exception("Cannot operate Max")
        Max = max(temp)
        self.res.push(Max)    #push back the maximum
        
    def average(self): #if top list of res is consist of numbers, then operate happens
        temp = self.res.pop()
        Sum = 0
        for i in temp:
            if is_number(i):
                Sum += i
            else:
                raise Exception("Cannot operate average")
        ave = int(Sum / len(temp))
        self.res.push(ave)    #push back the average number   
        
    def count(self):  #count numbers in top list e.g.[Flora, Hama]
        content = self.res.pop()
        self.res.push(len(content))         

############################################################################################
    def Exist(self): #whether elements in top exists in object_store
        temp = self.res.pop()
        for i in temp:
            if is_number(i):
                if i > self.Object_store.store[-1][-1][-1][-1][0]:
                    self.res.push(False)
                    return
                
                else:                
                    self.res.push(True)
                    return
    def In(self): #check whether x is in Q or not. Q is the top stack and x is the input value of user
        #e.g. x = Flora Q = [i3,i5,i6,i22]
        temps = self.res.pop() #lager one(ideally)
        #self.res.dereference()  #change ids to values
        tempb = self.res.pop() #smaller one(ideally) ["Flora"]
        count = 1
        for i in temps:
            self.res.push(self.nested(i))     
            self.dereference()
            for k in tempb:
                if k in self.res.top:  
                    count = 0
                else:
                    count=1
            self.res.pop()
            if count==0:        
                self.res.push(True)
                return
        self.res.push(False)
        return
                    
    def Contains(self): #Q1 pop first, Q2 pop next, check whether Q1 contains in Q2
        tempb = self.res.pop() #smaller one(ideally)
        temps = self.res.pop() #larger one(ideally)
        if len(tempb) < 1:
            raise Exception("Empty stack")
        for i in temps:
            if type(tempb)==list:
                if i == tempb[0]:
                    self.res.push(True)
                    return
            elif type(tempb)==int:
                if i == tempb:
                    self.res.push(True)
                    return
            elif type(tempb)==str:
                if i == tempb:
                    self.res.push(True)
                    return
        self.res.push(False)
    def distinct(self):
        m=self.res.pop()
        mergeSort(m,0,len(m)-1)
        i = 0
        while i < (len(m) - 1):
            if m[i]==m[i+1]:
                m.pop(i+1)
            else:
                i += 1
        self.res.push(m)   
############################################################################################   
    def dereference(self):    #for res(top),return should be a list containing all elements without ids
        temp1 = self.res.pop()
        len_temp1 = len(temp1) 
        result = []   #result list, empty initially
        for i in temp1: #temp1 might contains many ids.
            if is_number(i) == False:
                raise Exception("Already dereferenced.")
            if i > self.Object_store.store[-1][-1][-1][-1][0]:
                raise Exception("It's out of range")
            j = len(self.Object_store.store)#level 1: The big class of 3 Theaters
            #initially, find the id!!!
            while j > 0: #j is j(th)
                if i < self.Object_store.store[j-1][0][0]:
                    j -= 1
                else:
                    break #we find the number in the first level
            k = len(self.Object_store.store[j-1])
            while k > 0:
                if i == self.Object_store.store[j-1][k-1][0]:
                    raise Exception("id cannot be dereferenced")
                elif i<self.Object_store.store[j-1][k-1][0]:
                    k -= 1
                else:
                    break #find!
            delta = i - self.Object_store.store[j-1][k-1][0] - 1
            ans = self.Object_store.store[j-1][k-1][2][delta][2]
            result.append(ans)
                    #for k in j: #level 2: the single class of Theater
                        #if k[0] ==temp1[i]
                        #res.append(k[2])
            self.res.push(result)
    
    def nested(self,i):   #tell if i is the first level or in the 2 level
        result = []
        for j in self.Object_store.store:
            for k in j:
                if i == k[0]:
                    if type(k[2])==int:
                        return None
                    else:
                        for m in k[2]:
                            result.append(m[0])
                    break
        return result
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass
 
    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass
 
    return False

def sort_stack(stack):
    li=[]
    while stack.is_Empty()==False:
        li.stack(self.pop)
    
    mergeSort(li,0,len(li))
    for i in li:
        stack.p
    
def merge(arr, l, m, r): 
    n1 = m - l + 1
    n2 = r- m 
  
    # temporary array
    L = [0] * (n1)
    R = [0] * (n2)
  
    # copy data to arrays L[] and R[] 
    for i in range(0 , n1): 
        L[i] = arr[l + i] 
  
    for j in range(0 , n2): 
        R[j] = arr[m + 1 + j] 
  
    # initialize some variables
    i = 0  
    j = 0 
    k = l  
  
    while i < n1 and j < n2 : 
        if L[i] <= R[j]: 
            arr[k] = L[i] 
            i += 1
        else: 
            arr[k] = R[j] 
            j += 1
        k += 1
  
    #  copy the remaining of L[]
    while i < n1: 
        arr[k] = L[i] 
        i += 1
        k += 1
  
    # copy the remaining of R[]
    while j < n2: 
        arr[k] = R[j] 
        j += 1
        k += 1
        
def mergeSort(arr,l,r): 
    if l < r:         
        m = int((l+(r-1))/2)
        mergeSort(arr, l, m) 
        mergeSort(arr, m+1, r) 
        merge(arr, l, m, r) 
```

## Operation Rule of the Two stack machine
###### The object store is to contain a finite, non-empty collection of named lists L1, . . . , Lk.
###### ENV denotes the environment stack that is used to capture bindings of variables.
###### RES denotes the result stack that is to capture intermediate and final results.
       1.Queries are defined inductively as follows:
           – Each list name Li in the object store is a query;
           – If Q is a query and ϕ is a selection formula applicable to the result of Q, then Q where ϕ is a query;
           – If Q is a query and A is an attribute of the result of Q, then Q.A is a query;
           – For queries Q1 and Q2 the cartesian product Q1 × Q2 is a query;
           – Nothing else is a query.
       2.For each operator or operand, we use '(' and ')' to indicate the prority. 
       3.For each operator or operand, we need spacing to distinguish and seperate them.
       4.The sequence of operations is broken up from left to right and passed using recursion.
       5.The operation are further divided into {where, . , x} ,two side algebraic requests  Q1ΘQ2 and one side algebraic requests  ΘQ
            - self.command=["where","x","."]
            - self.one_command=["count", "sum", "min", "max", "average", "distinct", "not" , "deref","nested"]
            - self.two_command=["=", "<", ">", "<=", ">=", "neq","and", "or","+", "-", "*","/","Exists","Contains", "In"]
       6.For efficency, equaljoin and distinct are using mergesort method to improve the time complexity.
       7.We also define many specific cases for add into three categories:
       
           -one number plus another number 
           -two list concat together
           -one element (ie,single element array) with a number
       8.For "=", "<", ">", "<=", ">=", "neq" use the python internal function, only elements with samee type can be compared and, for those with different types we will have error.
       9.For dereference operation, we can't apply it to the List name but only to thee attribut name and we can get the corresponding value.
